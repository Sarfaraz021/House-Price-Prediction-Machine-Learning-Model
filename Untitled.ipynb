```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
model = linear_model.LinearRegression()
```

```python
x = final.drop('price', axis  = 'columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() #Creating Class Instance/Object
```

```python
dfle = df
```

```python
dfle.town = le.fit_transform(dfle.town) # this method will create labels of specied column
```

```python
dfle
```

```python
X = dfle[['town','area']].values # we used  value to make it 2d array instead dataframe (.values will return 2d array)
X
```

```python
Y = dfle.price
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe = OneHotEncoder()
```

```python
column_transformer = ColumnTransformer([('encoder', ohe, [0])],
                
                                       remainder='passthrough')
```

```python
X = column_transformer.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[1,0,2800]])
```
